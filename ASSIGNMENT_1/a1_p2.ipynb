Visualization of Data using matplotlib, pyplots packages

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# generating data sets
people = np.array(["kiran", "arun", "vijay", "varun"])
age = np.array([25, 30, 35, 40])
height = np.array([145, 151, 165, 173])
weight = np.array([45, 55, 65, 75])
```

```python
# 1. Scatter Plot: Height vs Weight
"""
    Justification:
    This scatter plot represents the relationship between height and weight of the individuals.
    As height increases, weight tends to increase as well, indicating a positive correlation.
"""
plt.figure(figsize=(10, 6))
plt.scatter(height, weight, color='blue', marker='o')
plt.title('Scatter Plot: Height vs Weight', fontsize=16)
plt.xlabel('Height (cm)', fontsize=14)
plt.ylabel('Weight (kg)', fontsize=14)
plt.grid(True)
plt.show()
```

```python
# 2. Bar Chart: Age vs Weight
"""
  Justification:
  The bar chart shows the weight of each individual by age. It is clear that as age increases,
  weight also increases, which might suggest that as people age, their weight tends to rise,
  though this is a small dataset.
"""
plt.figure(figsize=(10, 6))
plt.bar(people, weight, color='green')
plt.title('Bar Chart: Age vs Weight', fontsize=16)
plt.xlabel('People', fontsize=14)
plt.ylabel('Weight (kg)', fontsize=14)
plt.show()
```

```python
# 3. Histogram: Distribution of Heights
"""
  Justification:
  The histogram represents the distribution of heights in the dataset.
  It shows how many people fall into specific height ranges. In this dataset,
  most people are clustered in the middle height range.
"""
plt.figure(figsize=(10, 6))
plt.hist(height, bins=5, color='purple', edgecolor='black')
plt.title('Histogram: Distribution of Heights', fontsize=16)
plt.xlabel('Height (cm)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.show()
```
